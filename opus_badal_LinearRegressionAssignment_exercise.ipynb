### 1. What is Regression?

Regression is a statistical method used to understand the relationship between a dependent variable and one or more independent variables. It is mainly used to predict continuous values such as house prices, sales, or temperature.

### 2. What is Linear Regression?

Linear Regression is a supervised learning algorithm that models the relationship between input variables and a continuous output variable using a straight line.

### 3. When to use Linear Regression? Explain the equation of a straight line.

Linear regression is used when the target variable is continuous and the relationship between variables is roughly linear. The equation of a straight line is y = mx + c where m is the slope and c is the intercept.

### 4. What kind of plots will you use to showcase the relationship amongst the columns?

Scatter plots are commonly used to visualize relationships between variables. Pair plots and correlation heatmaps can also help understand relationships between multiple columns.

### 5. How is the best fit line chosen?

The best fit line is chosen by minimizing the error between predicted values and actual values. This is usually done using the least squares method which minimizes the squared residuals.

### 6. What is gradient descent, and why is it used? Explain the maths.

Gradient descent is an optimization algorithm used to minimize the loss function. It updates model parameters step by step in the direction that reduces the error.

### 7. What are residuals?

Residuals are the differences between actual values and predicted values produced by the regression model.

### 8. What is correlation?

Correlation measures the strength and direction of the relationship between two variables. Its value ranges from -1 to 1.

### 9. What is multicollinearity?

Multicollinearity occurs when two or more independent variables are highly correlated with each other which can affect the reliability of regression coefficients.

### 10. How to detect multicollinearity?

Multicollinearity can be detected using a correlation matrix, pair plots, or Variance Inflation Factor (VIF).

### 11. What are the remedies for multicollinearity?

It can be handled by removing highly correlated features, using dimensionality reduction, or applying regularization techniques.

### 12. What is the R-Squared Statistics?

R-squared measures how well the regression model explains the variation in the dependent variable.

### 13. What is an adjusted R-Squared Statistics?

Adjusted R-squared is a modified version of R-squared that adjusts for the number of independent variables in the model.

### 14. Why do we use adj R-squared?

We use adjusted R-squared because it penalizes the addition of unnecessary variables and provides a more reliable evaluation.

### 15. Why adj R-squared decreases when we use incompetent variables?

If a variable does not improve the model significantly, adjusted R-squared may decrease because it penalizes useless predictors.

### 16. How to interpret a Linear Regression model?

A regression model is interpreted by looking at coefficients which show how much the dependent variable changes when an independent variable increases by one unit.

### 17. What is the difference between fit, fit_transform and predict methods?

fit trains the model, fit_transform trains and transforms data, and predict generates predictions using the trained model.

### 18. How do you plot the least squared line?

The least squares line can be plotted by creating a scatter plot of the data and drawing the regression line that represents predicted values.

### 19. What are Bias and Variance? What is Bias Variance Trade-off?

Bias is error from overly simple models while variance is error from models being too sensitive to training data. The trade-off is balancing both.

### 20. What is the null and alternate hypothesis?

The null hypothesis states there is no relationship between variables while the alternative hypothesis states that a relationship exists.

### 21. What is multiple linear regression?

Multiple linear regression models the relationship between one dependent variable and multiple independent variables.

### 22. What is the OLS method?

OLS (Ordinary Least Squares) estimates regression parameters by minimizing the sum of squared residuals.

### 23. What is the p-value? How does it help in feature selection?

The p-value helps determine whether a feature is statistically significant. Lower p-values indicate stronger evidence that a variable is useful.

### 24. How to handle categorical values in the data?

Categorical variables can be handled using encoding techniques such as one-hot encoding or label encoding.

### 25. What is regularization, and why do we need it?

Regularization helps prevent overfitting by adding a penalty term to the loss function.

### 26. Explain Ridge Regression.

Ridge regression adds an L2 penalty to shrink coefficient values and reduce model complexity.

### 27. Explain Lasso Regression.

Lasso regression adds an L1 penalty which can shrink some coefficients to zero and perform feature selection.

### 28. Explain Elastic Net.

Elastic Net combines both L1 and L2 penalties from Lasso and Ridge regression.

### 29. Why do we do a train test split?

Train-test split helps evaluate model performance on unseen data and prevents overfitting.

### 30. What is polynomial regression? When to use it?

Polynomial regression models non-linear relationships by including polynomial terms of the predictors.

### 31. Explain the steps for GCP deployment.

Steps include training the model, saving the model, creating a cloud instance, deploying the API, and testing the service.

### 32. What difficulties did you face in cloud deployment?

Common issues include dependency conflicts, configuration problems, and managing cloud resources.
